# --- üìå Operators in Python ---

Operators are special symbols in Python used to perform operations on variables and values.

---

### üîπ 1. Arithmetic Operators
Used to perform basic mathematical operations:

| Operator | Description      | Example    | Output |
|----------|------------------|------------|--------|
| `+`      | Addition          | `3 + 2`    | `5`    |
| `-`      | Subtraction       | `3 - 2`    | `1`    |
| `*`      | Multiplication    | `3 * 2`    | `6`    |
| `/`      | Division          | `3 / 2`    | `1.5`  |
| `//`     | Floor Division    | `3 // 2`   | `1`    |
| `%`      | Modulus (remainder)| `3 % 2`   | `1`    |
| `**`     | Exponentiation    | `3 ** 2`   | `9`    |

---

### üîπ 2. Comparison (Relational) Operators
Used to compare two values and return `True` or `False`:

| Operator | Description     | Example     | Output   |
|----------|------------------|-------------|----------|
| `==`     | Equal to          | `3 == 2`    | `False`  |
| `!=`     | Not equal to      | `3 != 2`    | `True`   |
| `>`      | Greater than      | `3 > 2`     | `True`   |

| `>=`     | Greater or equal  | `3 >= 2`    | `True`   |
| `<=`     | Less or equal     | `3 <= 2`    | `False`  |

---

### üîπ 3. Logical Operators
Used to combine conditional statements:

| Operator | Description               | Example                 | Output   |
|----------|----------------------------|--------------------------|----------|
| `and`    | True if both are True      | `True and False`         | `False`  |
| `or`     | True if one is True        | `True or False`          | `True`   |
| `not`    | Reverse the result         | `not(True)`              | `False`  |

---

### üîπ 4. Assignment Operators
Used to assign values to variables:

| Operator | Example  | Equivalent To   |
|----------|----------|-----------------|
| `=`      | `x = 5`  | Assigns 5 to x  |
| `+=`     | `x += 2` | `x = x + 2`     |

Check whether a value is in a sequence (like a list, tuple, string, etc.):

| `in`     | True if found in sequence  | `'a' in 'cat'`   | `True`   |

| Operator | Description            | Example      | Output   |

````python
from rdkit import Chem
from rdkit.Chem.Descriptors import MolWt

def get_molecular_weight(smiles):
  """Calculates the molecular weight of a molecule from its SMILES string."""
  mol = Chem.MolFromSmiles(smiles)
  if mol:
    return MolWt(mol)
  else:
    return None

# Example usage:
aspirin_smiles = "CC(=O)OC1=CC=CC=C1C(=O)O"
mol_weight = get_molecular_weight(aspirin_smiles)

if mol_weight is not None:
  print(f"The molecular weight of Aspirin ({aspirin_smiles}) is: {mol_weight:.2f}")
else:
  print(f"Could not calculate molecular weight for {aspirin_smiles}")
````

````python
# Generate chemoinformatics molecule information data
from rdkit import Chem
from rdkit.Chem.Descriptors import MolWt
import pandas as pd

smiles_list = [
    "CCO",                      # Ethanol
    "CC(=O)OC1=CC=CC=C1C(=O)O", # Aspirin
    "C1=CC=CC=C1",               # Benzene
    "CN",                       # Methylamine
    "CC(=O)N",                   # Acetamide
    "C",                        # Methane
    "CO",                       # Methanol
    "CCN"                       # Ethylamine
]

molecule_info_data = []
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        mw = MolWt(mol)
        molecule_info_data.append({"SMILES": smiles, "Molecule": mol, "MolecularWeight": mw})
    else:
        print(f"Could not process SMILES: {smiles}")

df_molecule_info = pd.DataFrame(molecule_info_data)
print("Chemoinformatics Molecule Information Data:")
display(df_molecule_info)
````

## ü§ñ Section 2: Introduction to Scikit-learn
Now that you know Python basics, let's try a simple machine learning example!

We will use the **Iris flower dataset**, which contains measurements of different types of iris flowers. The goal is to classify the flower species based on these measurements.

### üîπ Install scikit-learn
Run this cell only once to install the library.

### üîπ Import Required Libraries

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
````

### üîπ Load the Dataset
The `load_iris()` function gives us features (measurements) and labels (species name) for iris flowers.

````python
iris = load_iris()
print("Feature names:", iris.feature_names)
print("Target names:", iris.target_names)
````

````python
import pandas as pd

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
display(iris_df.head())
````

### üîπ Split the Dataset
We divide the data into training (80%) and testing (20%) sets to evaluate model performance.

````python
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.2, random_state=42
)

print("First training sample:", X_train[0])
print("Its label:", y_train[0])
````

image (1).png
### üîπ Train a Machine Learning Model
We‚Äôll use the **K-Nearest Neighbors (KNN)** algorithm. It classifies a data point based on how its neighbors are classified.

````python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
````

# --- üå≤ Random Forest Classifier ---

### üîπ What is a Random Forest?

A **Random Forest** is an **ensemble learning method** that builds multiple decision trees and combines their results to make a more accurate and stable prediction.

- It works for **classification** (categorical labels) and **regression** (continuous values).
- It reduces **overfitting** by averaging the predictions of many decision trees.
image (3).png
---

````python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
````

# --- üìà Support Vector Machine (SVM) Classifier ---

### üîπ What is an SVM?

A **Support Vector Machine (SVM)** is a supervised learning algorithm used for **classification** and **regression** tasks. It tries to find the **best decision boundary (hyperplane)** that separates classes in your feature space.

- Works well in **high-dimensional** spaces
- Particularly effective in cases where the number of dimensions > number of samples
- Commonly used in **bioinformatics** and **cheminformatics** for molecule classification

---image (4).png

````python
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
````

### üîπ Make Predictions and Evaluate
We use the trained model to make predictions on the test data and compare it to the actual results.

````python
y_pred = model.predict(X_test)
print("Predictions:", y_pred)
print("Actual:", y_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
````

````python
y_pred = rf_model.predict(X_test)
print("Predictions:", y_pred)
print("Actual:", y_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
````

````python
y_pred = svm_model.predict(X_test)
print("Predictions:", y_pred)
print("Actual:", y_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
````

### üîπ Make a Custom Prediction
We predict the class of a new sample with specific flower measurements.

````python
sample = [[5.1, 3.5, 1.4, 0.2],[3.3,2.1,1.9,0.9]]
predicted_class = model.predict(sample)
print("Predicted class index:", predicted_class)
print("Predicted flower:", iris.target_names[predicted_class[1]])
print("Predicted flower:", iris.target_names[predicted_class])
````
